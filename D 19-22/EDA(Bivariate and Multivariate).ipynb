## 5.HeatMap(categorical-Categorical)

```python
titanic=pd.read_csv('train.csv')
titanic.head(2)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic1=titanic.iloc[:,[0,1,2,5,9,4,11]]
titanic1
```

```python
titanic1.groupby('Pclass').mean()['Survived']*100
```

```python
(titanic1.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
(titanic1.groupby('Sex').mean()['Survived']*100).plot(kind='bar')
```
